# 1: Linear Regression - Boston Housing Price Prediction

### Task
- Load the Boston Housing dataset
- Preprocess the data
- Split into training and testing sets
- Train a Linear Regression model
- Evaluate model performance

```python
# Load Boston Housing dataset
boston = fetch_openml(name="boston", version=1, as_frame=True)

# Extract features and target
X, y = boston.data, boston.target
```

```python
# TODO: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=38)
```

```python
len(X_train), len(y_train),len(X_test),len(y_test)
```

```python
# TODO: Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# TODO: Train a Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)
```

```python
# TODO: Make predictions and calculate performance metrics
y_pred = lr_model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
```

```python
# TODO: Print performance metrics
print(f"Mean Squared Error: {mse:.2f}")
```

## Part 3: Decision Tree - Titanic Survival Prediction

### Task
- Load the Titanic dataset
- Preprocess the data (handling missing values, encoding categorical variables)
- Split into training and testing sets
- Train a Decision Tree Classifier
- Evaluate model performance

```python
# Load Titanic dataset
titanic = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
```

```python
# Data Preprocessing
# TODO: Handle missing values
titanic.info()
```

```python
titanic.tail()
```

```python
missing_values = titanic.isnull().sum()
print(f"Missing values per column:\n{missing_values}")
```

```python
categorical_columns = ['Survived', 'Pclass','Sex','Age','Siblings/Spouses Aboard','Parents/Children Aboard','Fare']
for col in categorical_columns:
     print(f"Category in {col} is : {titanic[col].unique()}\n")
```

```python
df=titanic.copy()
```

```python
# TODO: Encode the features
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df.describe()
```

```python
# Feature selection
X_titanic = df[['Pclass', 'Sex', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare']]
y_titanic = df['Survived']
```

```python
X_titanic, y_titanic
```

```python
# TODO: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_titanic, y_titanic, test_size=0.2, random_state=42)
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
# TODO: Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
parameters = [1,2,3,4,5,6,7,8,9,13]
for parameter in parameters:
    tree_model = DecisionTreeClassifier(max_depth=parameter, random_state=42)
    tree_model.fit(X_train, y_train)
    
    y_pred = tree_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"Accuracy: {accuracy:.2f} for parameter: {parameter}")
```

```python
# TODO: Train a Decision Tree Classifier
tree_model = DecisionTreeClassifier(max_depth=8, random_state=42)
tree_model.fit(X_train, y_train)
```

```python
# TODO: Make predictions and calculate performance metrics
y_pred = tree_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
# TODO: Print performance metrics
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# TODO: Visualize Decisoion Tree
plt.figure(figsize=(20, 15))
plot_tree(tree_model, feature_names=X_titanic.columns, class_names=['Not Survived' ,'Survived'], filled=True)
plt.title('Decision Tree for Titanic Survival Classification')
plt.show()
#plt.savefig('decision_tree_titanic.png') #la salvo perchè non si vede bene
```
